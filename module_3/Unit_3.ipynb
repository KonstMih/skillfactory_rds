## Составим список кухонь

```python
df['Cuisine Style'] = df['Cuisine Style'].fillna(1)
```

```python
df.info()
```

```python
kuhni_spisok = set()
```

```python
def kuhni(i):
    if i != 1:
        result = re.findall(r'[^\"\[\]\'\, ][\w|\s&]+', i)
        for n in result:
            kuhni_spisok.add(n)
```

```python
df['Cuisine Style'].apply(kuhni)
```

```python
len(kuhni_spisok)
```

```python
for i in kuhni_spisok:
    df[i] = df['Cuisine Style'].apply(lambda x: 1 if ((i in str(x))) else 0)
```

```python
df
```

```python
df['Cuisine Style'] = df['Cuisine Style'].apply(lambda x: len(re.findall(r'[^\"\[\]\'\, ][\w|\s&]+', x)) if x!=1 else x)
```

```python
def cena(x):
    if x=='$':
        return 1
    elif x=='$$ - $$$':
        return 2
    elif x=='$$$$':
        return 3
    else:
        return x
```

```python
df['Price Range'] = df['Price Range'].apply(cena)
```

```python
df['Price Range'] = df['Price Range'].fillna(2)
```

```python
def dni(x):
    result = re.findall(r'\d\d/\d\d/\d\d\d\d', x)
    if len(result) == 2:
        n = np.abs(datetime.strptime(result[0], '%m/%d/%Y') - datetime.strptime(result[1], '%m/%d/%Y'))
        return n
    elif len(result) == 1:
        return 0
    elif len(result) == 0:
        return -1
```

```python
df['Reviews'] = df['Reviews'].apply(dni)
```

```python
df = df.drop(['City', 'URL_TA', 'ID_TA'], axis = 1)
```

```python
df['Reviews'] = df['Reviews'].apply(lambda x: x.days if (x!=0 and x!=-1) else x)
```

```python
df['Number of Reviews'] = df['Number of Reviews'].fillna(0)
```

```python
df
```

# Разбиваем датафрейм на части, необходимые для обучения и тестирования модели

```python
# Х - данные с информацией о ресторанах, у - целевая переменная (рейтинги ресторанов)
X = df.drop(['Restaurant_id', 'Rating'], axis = 1)
y = df['Rating']
```

```python
# Загружаем специальный инструмент для разбивки:
from sklearn.model_selection import train_test_split
```

```python
# Наборы данных с меткой "train" будут использоваться для обучения модели, "test" - для тестирования.
# Для тестирования мы будем использовать 25% от исходного датасета.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

# Создаём, обучаем и тестируем модель

```python
# Импортируем необходимые библиотеки:
from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели
from sklearn import metrics # инструменты для оценки точности модели
```

```python
# Создаём модель
regr = RandomForestRegressor(n_estimators=100)

# Обучаем модель на тестовом наборе данных
regr.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = regr.predict(X_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```
